```python
#5.Pan numarası boş olan satırları bul ve kaldır.
```

```python
df = df.dropna(subset=['Pan'])
df = df[df['Pan'].str.strip() != '']

print(df)
```

```python
#6.Aynı PAN numarasına sahip satırlar arasında PaymentTransactionID'si farklı olanları belirle.
```

```python
# PAN numarasına göre gruplama
grouped = df.groupby('Pan')

# Filtrelenmiş veriyi saklamak için bir liste oluşturma
filtered_data = []

# Her grubu görüntüleme
columns_to_display = ['PaymentTransactionID', 'Pan', 'TransactionType', 'CreateDate', 'TransactionStatus', 'Amount', 'TransactionRequestDate', 'TransactionResponseDate']

for pan, group in grouped:
    # Farklı PaymentTransactionID sayısını kontrol etme
    if group['PaymentTransactionID'].nunique() >= 2:
        print(f"PAN: {pan}")
        print(group[columns_to_display])
        print()
```

```python
# PAN numarasına göre gruplama
grouped = df.groupby('Pan')

# Filtrelenmiş veri saklama
filtered_data = []

# Her grubu kontrol etme ve veriyi toplama
for pan, group in grouped:
    if group['PaymentTransactionID'].nunique() >= 2:
        filtered_data.append((pan, group))

# Grafik düzeni ayarlama
num_plots = len(filtered_data)
cols = 3  # Her satırda 3 grafik
rows = (num_plots // cols) + (num_plots % cols > 0)  # Satır sayısını hesaplama

fig, axes = plt.subplots(rows, cols, figsize=(15, 5 * rows), constrained_layout=True)

# Eğer sadece bir satır varsa axes'i düzleştir
if rows == 1:
    axes = axes.reshape(1, cols)

# Her grup için grafik oluşturma
for ax, (pan, group) in zip(axes.flatten(), filtered_data):
    sns.scatterplot(data=group, x='CreateDate', y='Amount', hue='PaymentTransactionID', palette='viridis', ax=ax)
    ax.set_title(f'Amount vs. CreateDate for PAN: {pan}')
    ax.set_xlabel('CreateDate')
    ax.set_ylabel('Amount')
    ax.legend(title='PaymentTransactionID')
    ax.tick_params(axis='x', rotation=45)

# Boş olan eksenleri gizleme (eğer varsa)
for i in range(len(filtered_data), len(axes.flatten())):
    fig.delaxes(axes.flatten()[i])

# Grafikleri gösterme
plt.show()
```

```python
#5.CreateDate sütununda tarih ve saatleri aynı olan fakat dakikaları arasındaki fark 1 saatten az olan satırları belirle.
```

```python
# PAN numarasına göre gruplama
def filter_time_diffs(group):
    # Verileri sıralama
    group = group.sort_values(by='CreateDate')
    # Farkları hesaplama
    group['PrevDate'] = group['CreateDate'].shift(1)
    group['Difference'] = (group['CreateDate'] - group['PrevDate']).dt.total_seconds() / 3600  # saat cinsinden fark
    # 1 saatten az olanları filtreleme
    return group[group['Difference'] < 1]

# Her grup için filtreleme işlemi, indekslerin korunmasını sağlama
filtered_df = df.groupby('Pan', group_keys=False).apply(filter_time_diffs)

# Sonuçları görüntüleme
print(filtered_df)
```
